```python
import pandas as pd
```

```python
df = pd.read_csv('universal_top_spotify_songs.csv', engine='python', on_bad_lines='skip')
```

```python
print(df.head(10))
```

```python
print(df.columns.tolist())
```

```python
df['snapshot_date'] = pd.to_datetime(df['snapshot_date'], errors='coerce')
df['album_release_date'] = pd.to_datetime(df['album_release_date'], errors='coerce')
```

```python
print(df[['name', 'artists', 'album_release_date']].head(10))
```

```python

problem_rows = df[pd.to_datetime(df['album_release_date'], errors='coerce').isna() & 
                 df['album_release_date'].notna()]

print(f"Found {len(problem_rows)} rows where album_release_date looks like text:")
print(problem_rows[['name', 'artists', 'album_release_date']].head(10))
```
